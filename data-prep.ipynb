```python
def find_mukim_postcode_matches(text, reference_df):
    matches = set()  # Use set directly to avoid duplicates
    for idx, row in reference_df.iterrows():
        if (row['state'].lower() in text.lower()) and (row['district'].lower() in text.lower()) and (row['location'].lower() in text.lower()):
            matches.add((row['postcode']))
    return matches  # Returns set of tuples containing (postcode, state)
 
 
```

```python
mukim_district_state['address_partial'] = (mukim_district_state['mukim'].astype(str)+','+mukim_district_state['district'].astype(str)+','+mukim_district_state['state'].astype(str)).apply(lambda x: find_mukim_postcode_matches(x, location)).to_list()
#mukim_district_state['address_partial'] = (mukim_district_state['mukim'].astype(str)).apply(lambda x: find_mukim_postcode_matches(x, location['location'])).to_list()
mukim_district_state['address_partial'] = mukim_district_state['address_partial'].astype(str).str.replace(r"[\{\}\(\)]|set|'", '', regex=True).str.split(',')
mukim_district_state['postcode'] = mukim_district_state['address_partial'].apply(lambda x: x[0])
mukim_district_state
```
